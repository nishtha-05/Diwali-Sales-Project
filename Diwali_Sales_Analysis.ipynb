```python
import numpy as np
import pandas as pd # help to work on dataframe or tables
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
                                                          -- # Data Cleaning --
```

```python
df = pd.read_csv('Diwali_dataset.csv')
print(df)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

#                                          -- Exploratory Data Analysis --

#--Gender--

```python
ax=sns.countplot(x='Gender', data=df ,color='green')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x= 'Gender', y='Amount',data= sales_gen)
```

#                                                              Age

```python
ax = sns.countplot(data=df , x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_age = df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
df.head()
```

```python
top_states = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=top_states , x= 'State',y='Orders',color='r')
```

```python
top_purchase = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=top_purchase , x= 'State',y='Amount',color='g')
```

#                                                        Marital Status	

```python
ax= sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status' ,y='Amount', data=sales_status, hue='Gender')
```

#                                                           Occuption

```python
df['Occupation'].nunique()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

#                                                            Product Category

```python
sns.set(rc={'figure.figsize':(27,5)})
ax=sns.countplot(x='Product_Category',data=df,color='g')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category' ,y='Amount', data=sales_status)
```

#                                                     Product id

```python
sales_status = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID' ,y='Orders', data=sales_status)
```
